# 1. Probleemcontext en Maatschappelijke Relevantie
Binnen het hoger onderwijs is er veel vraag naar gepersonaliseerde ondersteuning bij het maken van studiegerelateerde keuzes. Studenten ervaren vaak keuzestress door het groot aantal verschillende mogelijkheden en vinden het ook lastig de beste keuze te maken, gelinkt aan hun interesses, waarden en/of lange-termijndoelen. Hierdoor worden keuzes regelmatig gebaseerd op oppervlakkige kenmerken (zoals gunstig rooster, vrienden). Dit kan dan weer leiden tot een lager studiesucces, verminderde motivatie en een mindere aansluiting bij de persoonlijke ontwikkeling van studenten. 

Het Lectoraat Digitale Didactiek wil verkennen hoe AI ingezet kan worden om studenten te helpen bewuster keuzes te maken in hun studiepad. Met specifiek de opdracht om een AI-powered Smart Study Coach te ontwikkelen die studenten aan hogeschool Avans gaat helpen met kiezen van passende VKM’s (Vrije Keuze Modules).

Niet alleen de studenten zelf, maar ook de maatschappij als geheel heeft baat bij studenten die meer bewuste keuzes maken: minder ‘switchen’ tussen studies, hoger gemotiveerde studenten en betere aansluitingen op de arbeidsmarkt. Bovendien zorgt het ontwikkelen van een AI-powered Smart Study Coach ervoor dat  keuzehulp  meer toegankelijk is voor alle studenten, ook degene die minder snel om hulp zouden vragen.
