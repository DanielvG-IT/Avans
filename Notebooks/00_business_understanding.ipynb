# 1. Probleemcontext en Maatschappelijke Relevantie
Binnen het hoger onderwijs is er veel vraag naar gepersonaliseerde ondersteuning bij het maken van studiegerelateerde keuzes. Studenten ervaren vaak keuzestress door het groot aantal verschillende mogelijkheden en vinden het ook lastig de beste keuze te maken, gelinkt aan hun interesses, waarden en/of lange-termijndoelen. Hierdoor worden keuzes regelmatig gebaseerd op oppervlakkige kenmerken (zoals gunstig rooster, vrienden). Dit kan dan weer leiden tot een lager studiesucces, verminderde motivatie en een mindere aansluiting bij de persoonlijke ontwikkeling van studenten. 

Het Lectoraat Digitale Didactiek wil verkennen hoe AI ingezet kan worden om studenten te helpen bewuster keuzes te maken in hun studiepad. Met specifiek de opdracht om een AI-powered Smart Study Coach te ontwikkelen die studenten aan hogeschool Avans gaat helpen met kiezen van passende VKM’s (Vrije Keuze Modules).

Niet alleen de studenten zelf, maar ook de maatschappij als geheel heeft baat bij studenten die meer bewuste keuzes maken: minder ‘switchen’ tussen studies, hoger gemotiveerde studenten en betere aansluitingen op de arbeidsmarkt. Bovendien zorgt het ontwikkelen van een AI-powered Smart Study Coach ervoor dat  keuzehulp  meer toegankelijk is voor alle studenten, ook degene die minder snel om hulp zouden vragen. 


## 2. Probleemstelling

Hoe kunnen studenten op een betrouwbare, transparante en persoonsgerichte manier ondersteund worden bij het kiezen van een passende Vrije Keuzemodule (VKM), door middel van een AI-gestuurd aanbevelingssysteem gebaseerd op hun interesses, waarden en doelen?

## 3. Juridische en Ethische Overwegingen (AVG & EU AI Act 2025)

### 3.1 Algemene Verordening Gegevensbescherming (AVG)

De gebruikte dataset (Uitgebreide_VKM_dataset.csv) bevat uitsluitend module-informatie en geen direct identificeerbare persoonsgegevens. Het studentprofiel dat door gebruikers wordt ingevoerd kan echter privacygevoelige informatie bevatten, zoals persoonlijke interesses of doelen. Daarom wordt het volgende toegepast:

- Dataminimalisatie: alleen data verzamelen die essentieel is voor het aanbevelingsmodel.
- Doelbinding: data wordt uitsluitend gebruikt voor het genereren van module-aanbevelingen.
- Transparantie: studenten begrijpen expliciet hoe hun gegevens worden verwerkt.
- Opslagbeperking: studentdata wordt niet langdurig opgeslagen of gedeeld buiten het prototype.
- Rechten van betrokkenen: studenten kunnen hun gegevens inzien, wijzigen of verwijderen.

### 3.2 Voorlopige Risicoclassificatie (EU AI Act 2025)

Volgens de EU AI Act 2025 valt een recommender-systeem voor onderwijsadvies niet onder de hoog-risico categorie, omdat het geen formele toegang tot onderwijs, examens of beoordelingen automatiseert. Het prototype wordt geclassificeerd als een limited-risk AI system. Hierdoor gelden de volgende verplichtingen:

- Transparantieplicht: gebruikers moeten weten dat zij met een AI-systeem werken.
- Documentatie en logging: inzicht in gebruikte data, algoritmes en beperkingen.
- Bias-controle: voorkomen dat bepaalde groepen systematisch benadeeld worden.
- Menselijke controle (Human-in-the-loop): het advies is altijd ondersteunend en ondergeschikt aan de vrije keuze van de student.

## 4. Verantwoord AI-ontwerp (Human-Centric Design)

Het project volgt de volgende human-centric AI-principes:

- Student houdt regie  
  Aanbevelingen zijn suggesties, geen beslissingen; de student blijft altijd eindverantwoordelijk.

- Uitlegbaarheid  
  Elke aanbeveling bevat een korte onderbouwing, bijvoorbeeld: “Deze module past bij jou omdat …”.

- Bias-mitigatie  
  Analyse en correctie van modulekenmerken en vectorrepresentaties om ongerechtvaardigde voorkeuren te verminderen.

- Veilige ontwikkeling  
  Dataset bevat geen persoonsgegevens; studentinput wordt lokaal en doelgebonden verwerkt.

- Inclusief ontwerp  
  Ondersteuning voor studenten met diverse achtergronden, leerstijlen en ervaringsniveaus.

## 5. Data Collection

### 5.1 Dataset

Voor de ontwikkeling van het AI-systeem wordt gebruikgemaakt van **Uitgebreide_VKM_dataset.csv**, aangeleverd door de opdrachtgever.

Deze dataset bevat onder andere:
- modulebeschrijvingen
- leerdoelen
- competenties
- categorieën/tags
- moeilijkheidsniveau
- themagebieden
- studielast
- vereiste voorkennis

De dataset is volledig anoniem en bevat uitsluitend kenmerken van modules, waardoor deze juridisch en ethisch geschikt is voor gebruik binnen het onderwijsdomein zonder aanvullende toestemming.

### 5.2 Data Workflow

- Data-inname: importeren van de dataset in de projectomgeving.
- Kwaliteitscontrole: controleren op ontbrekende waarden, inconsistenties en dubbele modules.
- Verrijking: extractie van tekstuele kenmerken met NLP-technieken.
- Veiligheid: studentprofielen worden lokaal gegenereerd en nooit gekoppeld aan identificeerbare gegevens.
